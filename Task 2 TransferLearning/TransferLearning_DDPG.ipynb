```python
import gymnasium as gym
import numpy as np
import matplotlib.pyplot as plt
from stable_baselines3 import DDPG
from stable_baselines3 import PPO
from stable_baselines3.common.noise import NormalActionNoise, OrnsteinUhlenbeckActionNoise
env = gym.make("Ant-v2")
n_actions = env.action_space.shape[-1]
action_noise = NormalActionNoise(mean=np.zeros(n_actions), sigma=0.1 * np.ones(n_actions))
model_DDPG = DDPG('MlpPolicy', env,
                policy_kwargs=dict(net_arch=[256, 256]),
                learning_rate=5e-4,
                buffer_size=15000,
                learning_starts=200,
                batch_size=32,
                gamma=0.8,
                train_freq=1,
                gradient_steps=1,
                verbose=1,
                tensorboard_log="ant/")
print("The Ant - v2  gyn environment and box object ")
model_DDPG.learn(int(100000))
model_DDPG.save("ddpg_ant")
env = model_DDPG.get_env()
model_DDPG = DDPG.load("ddpg_ant")
obs_old = env.reset()
obs_oldarr = np.array(obs_old)
print( obs_oldarr)
print("The shape of old is ",obs_oldarr.shape)
print("- - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - ") 
print("The Half Cheetah - v2  gym environment and box object ")
reward_arr = []
env1 = gym.make("HalfCheetah-v2")
obs_new = env1.reset()
obs_new=obs_new[0].tolist()
for i in range(17,111):
  obs_new.append(0)
obs_arr = np.matrix(obs_new)
print("The new environment for HalfCheetah is new one")
print(obs_arr)
print("The shape of new is ",obs_arr.shape)
print("- - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - ") 
done = False;
while  done != True :
    action, _states = model_DDPG.predict(obs_arr)
    print(action)
    obs_new, reward, term, info = env.step(action)
    reward_arr.append(reward)
    print(reward)
    done = term
plt.plot(np.cumsum(reward_arr)/ np.arange(1, len(reward_arr) + 1) )
plt.title("Average reward per episode")
plt.xlabel("Episode number")
plt.ylabel("Average reward")
plt.show()

```
